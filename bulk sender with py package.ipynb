```python
import pywhatkit
```

```python
start_hour = 17
start_time = 3
```

```python
with open('numbers.txt',encoding="utf8") as file:
    for index, line in enumerate(file):
        num = line.split(',')[0].strip()
        name = line.split(',')[1].strip()
        if '+' not in num:
            num = f"+{num}"
        text = f'''{name}, message'''      
        
        # send message here
        pywhatkit.sendwhatmsg(num, text, start_hour, start_time ,wait_time=10, tab_close=True)
    
        print(num)
        start_time += 1
        if start_time > 59:
            start_time = 0
            start_hour += 1
            if start_hour > 23:
                start_hour = 0
```
