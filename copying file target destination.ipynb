```python
import shutil

original = r'original_destionation'
target = r'target_destination'


lines = []
# txt includes files which we want to change its location
with open('listfile.txt', encoding="UTF-8") as f:
    lines = f.readlines()
```

```python
l = lines[0].split(',')
```

```python
for i in l:
    try:
        o = original+"/"+i
        shutil.copy(o, target)
    except: 
        pass
```
